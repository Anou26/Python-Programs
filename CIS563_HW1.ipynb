**4. Discretize attribute ‘how many’ into bins of size 10, e.g. 0-10, 11-20,. . .**

```python
x = data['how_many']
```

```python
def create_bins(lower_bound, width, quantity):   # This is a function to create multiple bins, since the no. of the bins required is high.
  bins2 = []
  for low in range(lower_bound, 
                  lower_bound + quantity*width + 1, width):
      bins2.append((low, low+width))
  return bins2
```

```python
bins2 = create_bins(lower_bound=0,            # Here, we're defining the start, interval_size (width) and the number of the bins(quantity).
                   width=10,
                   quantity=500)
bins3 = pd.IntervalIndex.from_tuples(bins2)   # Here, we're creating an index of intervals that are made of tuples closed on the same side.
x = data['how_many']
data['binned']=pd.cut(x,bins3)
```
